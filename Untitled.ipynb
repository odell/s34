```python
import brick
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy import stats
```

```python
import model_frequentist as model
```

```python
model.azr.config.n1, model.azr.config.n2
```

```python
model.norm_factor_penalty(model.azr.config.get_input_values())
```

```python
model.chi_squared(model.azr.config.get_input_values())
```

```python
theta0 = np.array(model.azr.config.get_input_values())
```

```python
theta1 = np.array([
    stats.norm(x, np.abs(x)/100).rvs() for x in theta0
])
```

```python
model.chi_squared(theta1)
```

```python
result = minimize(model.chi_squared, theta1)
```

```python
theta_star = result.x
```

```python
_, x1, x2, x3 = model.azr.predict(theta_star, dress_up=False)
```

```python
fig, ax = plt.subplots(dpi=200)
fig.patch.set_facecolor('white')

ax.plot(x1[:, 0], x2[:, 3] / x1[:, 3], marker='.', linestyle='')
ax.errorbar(x1[:, 0], x2[:, 5], yerr=x2[:, 6], linestyle='', capsize=3);
```

```python
fig, ax = plt.subplots(dpi=200)
fig.patch.set_facecolor('white')

ax.plot(x3[:, 0], x3[:, 4], marker='.', linestyle='')
ax.errorbar(x3[:, 0], x3[:, 7], yerr=x3[:, 8], linestyle='', capsize=3)

ax.set_xscale('log');
```
