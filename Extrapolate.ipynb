```python
import os
import numpy as np
import matplotlib.pyplot as plt
from corner import corner
import model_bayesian as model
import emcee
from multiprocess import Pool
from scipy import stats
```

```python
os.environ['OMP_NUM_THREADS'] = '1'
```

```python
flat_chain = np.load('model_1.npy')
```

```python
def s_factor(theta):
    return model.azr.extrapolate(theta, use_gsl=True, ext_capture_file='output/intEC.extrap')[0]
```

```python
with Pool(16) as pool:
    sfactors = pool.map(s_factor, flat_chain)
```

```python
fig, ax = plt.subplots(dpi=200)
fig.patch.set_facecolor('white')

ax.plot(sfactors[0][:, 0], sfactors[0][:, 4]);
```

```python
from scipy.optimize import curve_fit
```

```python
def fit(extrap):
    x = extrap[:, 0]
    y = extrap[:, 4]
    pars, cov = curve_fit(lambda z, m, b: m*z + b, x, y)
    return pars
```

```python
parameters = np.array([fit(x) for x in sfactors])
```

```python
fig, ax = plt.subplots(dpi=200)
fig.patch.set_facecolor('white')

ax.hist(parameters[:, 1]*1000, bins=50);
```

```python
low, med, high = np.quantile(parameters[:, 1]*1000, [0.16, 0.5, 0.84])
```

```python
med, high-med, med-low
```
